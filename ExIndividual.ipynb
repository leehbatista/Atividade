```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV baixado
filename = 'API_NY.GDP.PCAP.CD_DS2_en_csv_v2_571640.csv'
data = pd.read_csv(filename, skiprows=4)

# Verificar se os dados foram carregados corretamente
print(data.head())
```

```python
# Filtrar dados para alguns países específicos
countries = ['Brazil', 'China', 'United States', 'India']
data_filtered = data[data['Country Name'].isin(countries)]

# Configurar o índice
data_filtered.set_index('Country Name', inplace=True)
data_filtered = data_filtered.T[4:]  # Transpor e remover as primeiras linhas desnecessárias
data_filtered.columns = countries

# Verificar a estrutura dos dados filtrados
print(data_filtered.head())
```

Gráfico de Linha

```python
# Plotar o gráfico de linha
data_filtered.plot(figsize=(10, 6))
plt.title('PIB per capita (US$ corrente)')
plt.xlabel('Ano')
plt.ylabel('PIB per capita')
plt.legend(title='Países')
plt.grid(True)
plt.show()
```

Gráfico de Barra


```python
# Selecionar um ano específico para o gráfico de barra
data_2020 = data_filtered.loc['2020']

# Plotar o gráfico de barra
data_2020.plot(kind='bar', figsize=(10, 6))
plt.title('PIB per capita (US$ corrente) em 2020')
plt.xlabel('País')
plt.ylabel('PIB per capita')
plt.grid(True)
plt.show()
```

Histograma


```python
# Plotar histograma dos valores do PIB per capita em 2020
data_2020.plot(kind='hist', bins=10, figsize=(10, 6), alpha=0.7)
plt.title('Distribuição do PIB per capita (US$ corrente) em 2020')
plt.xlabel('PIB per capita')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

Gráfico de Pizza

```python
# Plotar gráfico de pizza para os valores do PIB per capita em 2020
data_2020.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Distribuição do PIB per capita (US$ corrente) em 2020')
plt.ylabel('')
plt.show()
```

Scatter Plot


```python
# Plotar scatter plot comparando PIB per capita de dois anos
plt.figure(figsize=(10, 6))
plt.scatter(data_filtered.loc['2010'], data_filtered.loc['2020'], c='blue', alpha=0.5)
plt.title('Comparação do PIB per capita (US$ corrente) entre 2010 e 2020')
plt.xlabel('PIB per capita em 2010')
plt.ylabel('PIB per capita em 2020')
plt.grid(True)
plt.show()
```
